# 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
excluded = df[(df['comments_disabled'] == True) & (df['ratings_disabled'] == True) | (df['video_error_or_removed'] == True)]
excluded
```

```python
df = df[~(((df['comments_disabled'] == True) & (df['ratings_disabled'] == True)) | (df['video_error_or_removed'] == True))]
df
```

# 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
df['like_ratio'] = df['likes']/df['dislikes']
```

```python
df
```

# 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df.publish_time=pd.to_datetime(df.publish_time, format='ISO8601')
```

```python
df
```

```python
start = df['publish_time'].dt.floor('10min')
end = start + pd.Timedelta(minutes=10)
df['time_cluster'] = start.dt.strftime('%H:%M') + ' - ' + end.dt.strftime('%H:%M')
```

```python
df
```

# 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
df.groupby(["time_cluster"])["video_id"].count()
```

```python
df.groupby(["time_cluster"])["likes"].mean()
```

```python
df.groupby(["time_cluster"])["dislikes"].mean()
```

```python
cluster_stats = df.groupby("time_cluster").agg(num_videos = ("video_id","count"), avg_likes = ("likes", "mean"), avg_dislikes = ("dislikes", "mean")).reset_index()
cluster_stats
```

# 8. For each tag, determine the number of videos. Notice that tags contains a string with several tags.

```python
tag_counts = df['tags'].str.split('|').explode().value_counts().reset_index()
tag_counts['tags']=tag_counts['tags'].str.strip().str.replace('"','',regex=False) # regex=False serve per trattare " come testo e non come pattern
tag_counts.columns = ['tags', 'num_videos']
tag_counts
```

```python
tag_counts = tag_counts.iloc[1:] # elimino la prima riga dell'indice perchè contiene un valore non desiderato
tag_counts
```

# 9. Find the tags with the largest number of videos

```python
tags_max_num_videos = int(tag_counts["num_videos"].max())
tags_max_num_videos
```

# 10. For each (tag, country) pair, compute average ratio likes/dislikes

```python
tag_counts_country = df[['tags', 'country', 'like_ratio']].copy()
tag_counts_country['tags'] = tag_counts_country['tags'].str.split('|')
tag_counts_country = tag_counts_country.explode('tags')
tag_counts_country['tags'] = tag_counts_country['tags'].str.strip().str.replace('"','',regex=False)
tag_counts_country = tag_counts_country[(tag_counts_country['tags'] != '[none]')&(tag_counts_country['tags'] != '')].reset_index(drop=True) #drop toglie indice ripetuto
tag_counts_country.groupby(["tags", "country"])[["like_ratio"]].mean()
```

# 11. For each (trending_date, country) pair, the video with the largest number of views

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df_date_country = df[['trending_date', 'country', 'views', 'video_id', 'title']].copy()
df7=df_date_country.groupby(['trending_date', 'country'])[['views']].max().reset_index()
es4=df_date_country.merge(df7, on=['trending_date', 'country', 'views'], how='inner').reset_index()
es4=es4.sort_values(by=['trending_date', 'country'])
es4.to_csv('es4.csv', index=False)
```

# 14. Read all json files with the video categories

```python
ca_json = pd.read_json('trendingYT/CA_category_id.json').assign(country='CA')
de_json = pd.read_json('trendingYT/DE_category_id.json').assign(country='DE')
fr_json = pd.read_json('trendingYT/FR_category_id.json').assign(country='FR')
gb_json = pd.read_json('trendingYT/GB_category_id.json').assign(country='GB')
in_json = pd.read_json('trendingYT/IN_category_id.json').assign(country='IN')
jp_json = pd.read_json('trendingYT/JP_category_id.json').assign(country='JP')
kr_json = pd.read_json('trendingYT/KR_category_id.json').assign(country='KR')
mx_json = pd.read_json('trendingYT/MX_category_id.json').assign(country='MX')
ru_json = pd.read_json('trendingYT/RU_category_id.json').assign(country='RU')
us_json = pd.read_json('trendingYT/US_category_id.json').assign(country='US')
df_json=pd.concat([ca_json,de_json,fr_json,gb_json,in_json,jp_json,kr_json,mx_json,ru_json,us_json], ignore_index=True)
df_json.to_csv('jsonconcat.csv', index=False)
df_json
```

```python
categories = df_json[['items', 'country']].copy()
categories['category_id'] = categories['items'].apply(lambda x: x['id'])
categories['category_title'] = categories['items'].apply(lambda x: x['snippet']['title'])
categories['assignable'] = categories['items'].apply(lambda x: x['snippet']['assignable'])
categories = categories[['category_id', 'category_title', 'assignable', 'country']].drop_duplicates().reset_index(drop=True)
categories.columns=['id_category', 'category_title', 'assignable', 'country']
categories
```

# 15. For each country, determine how many videos have a category that is not assignable.

```python
categories.id_category=categories.id_category.astype('Int64')
df_join_cat = df.merge(categories, how='inner', left_on='category_id' , right_on='id_category')
df_join_cat = df_join_cat.drop(columns=['id_category','country_y'])
df_join_cat
```

```python
not_assignable = df_join_cat[df_join_cat['assignable'] == False]
not_assignable
final_point = not_assignable.groupby("country_x")[['video_id']].aggregate('count').reset_index()
final_point.columns=['Country','Num_Videos']
final_point
```
